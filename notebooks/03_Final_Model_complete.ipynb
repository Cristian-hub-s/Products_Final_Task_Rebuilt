# 03 - Final Model

```python
import pandas as pd
import pickle

from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.svm import LinearSVC

# Load dataset
df = pd.read_csv('../data/products.csv')

# Clean column names
df.columns = df.columns.str.strip()

# Remove missing values
df = df.dropna(
    subset=[
        'Product Title',
        'Category Label'
    ]
)

# Remove rare categories
counts = df['Category Label'].value_counts()

valid_categories = counts[
    counts >= 2
].index

df = df[
    df['Category Label'].isin(valid_categories)
]

# Features and target
X = df['Product Title'].astype(str)

y = df['Category Label']

# Model
model = Pipeline([
    ('tfidf', TfidfVectorizer()),
    ('clf', LinearSVC())
])

# Train
model.fit(X, y)

# Save model
with open('../models/product_classifier.pkl', 'wb') as f:
    pickle.dump(model, f)

print("Model saved successfully.")
```
